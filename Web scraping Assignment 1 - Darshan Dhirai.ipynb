# Answer to Q2

```python
page2 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
page2
```

```python
soup2 = BeautifulSoup(page2.content)
name = []
rating = []
release_year = []
moviedata = soup2.find_all('h3', class_='lister-item-header')
md = soup2.find_all('div', class_='inline-block ratings-imdb-rating')
```

```python
for i in moviedata:
    for j in i.find_all('a'):
        name.append(j.text)
    for k in i.find_all('span', class_="lister-item-year text-muted unbold"):
        release_year.append(k.text)
```

```python
for i in md:
    for j in i.find_all('strong'):
        rating.append(i.text.replace('\n', ''))
```

```python
page22 = requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
soup22 = BeautifulSoup(page22.content)
moviedata1 = soup22.find_all('h3', class_='lister-item-header')
md1 = soup22.find_all('div', class_='inline-block ratings-imdb-rating')
for i in moviedata1:
    for j in i.find_all('a'):
        name.append(j.text)
    for k in i.find_all('span', class_="lister-item-year text-muted unbold"):
        release_year.append(k.text)
for i in md1:
    for j in i.find_all('strong'):
        rating.append(i.text.replace('\n', ''))
```

```python
imdbtop100 = pd.DataFrame({'Movie name': name, 'IMDB Rating': rating, 'Year of Release': release_year})
imdbtop100['rank'] = list(range(1,101))
imdbtop100.set_index(['rank'], drop= True, inplace=True)
imdbtop100
```

# Answer to Q4

```python
page4 = requests.get('https://bookpage.com/reviews')
page4
```

```python
soup4 = BeautifulSoup(page4.content)
bname = []
aname = []
gen = []
brev=[]
```

```python
no = 1
bn = soup4.find_all('h4', class_='italic')
for i in bn:
    if no<7:
        bname.append(i.text.replace('\n','').replace('★', '').strip())
        no+=1
    else:
        break
bname = bname[1:]
bname
```

```python
no1 = 1
an = soup4.find_all('p', class_='sans bold')
for i in an:
    if no1<7:
        aname.append(i.text.replace('\n', ''))
        no1+=1
    else:
        break
aname = aname[1:]
aname
```

```python
no2 = 1
gn = soup4.find_all('p', class_='genre-links hidden-phone')
for i in gn:
    if no2<7:
        gen.append(i.text.replace('\n', ''))
        no2+=1
    else:
        break
gen = gen[1:]
gen
```

```python
no3 = 1
rev = soup4.find_all('p', class_='excerpt')
for i in rev:
    if no3<6:
        if i.text !='\n':
            brev.append(i.text.replace('\n', ''))
            no3+=1
    else:
        break
brev
```

```python
bookreview = pd.DataFrame({'Book Name': bname, 'Author name': aname, 'Genre': gen, 'Review': brev, 'i': list(range(1,6))})
bookreview.set_index(['i'], drop= True, inplace=True)
bookreview
```

```python
page51 = requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page51
```

```python
soup5 = BeautifulSoup(page51.content)
trank = []
tname = []
tmatch = []
tpoint = []
trate = []
td1 = soup5.find_all('tr', class_='rankings-block__banner')
```

```python
for i in td1:
    trank.append(i.find('td', class_='rankings-block__banner--pos').text)
    tmatch.append(i.find('td', class_='rankings-block__banner--matches').text)
    tpoint.append(i.find('td', class_='rankings-block__banner--points').text)
    trate.append(i.find('td', class_='rankings-block__banner--rating u-text-right').text.replace('\n','').strip())
    for j in i.find_all('td', class_='rankings-block__banner--team-name'):
        tname.append(j.find('span', class_='u-hide-phablet').text)
```

```python
td2 = soup5.find_all('tr', class_='table-body')
```

```python
td2
```

```python
num = 1
for i in td2:
    trank.append(i.find('td', class_='table-body__cell table-body__cell--position u-text-right').text)
    for k in i.find_all('td', class_='table-body__cell u-center-text'):
        if num%2 == 1:
            tmatch.append(k.text)
            num +=1
        else:
            tpoint.append(k.text)
            num +=1
    trate.append(i.find('td', class_='table-body__cell u-text-right rating').text.replace('\n','').strip())
    for j in i.find_all('td', class_='table-body__cell rankings-table__team'):
        tname.append(j.find('span', class_='u-hide-phablet').text)
```

```python
oditeam = pd.DataFrame({'Rank': trank, 'Team': tname, 'Matches': tmatch, 'Points': tpoint, 'Rating': trate})
oditeam.set_index(['Rank'], drop= True, inplace=True)
oditeam
```

# Answer to Q8

```python
page8 = requests.get('https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.YQKeGj3itPY')
day =[]
sdes =[]
temp = []
desc = []
page8
```

```python
soup8 = BeautifulSoup(page8.content)
sday = soup8.find('div', id="seven-day-forecast")
fitems = sday.find_all('div', class_='tombstone-container')
for i in fitems:
    day.append(i.find('p', class_='period-name').text)
    sdes.append(i.find('p', class_='short-desc').text)
    temp.append(i.find('p', class_='temp').text)
```

```python
det = soup8.find_all('div', class_='col-sm-10 forecast-text')
ndet = 1
for i in det:
    if ndet <10:
        desc.append(i.text)
        ndet+=1
    else:
        break
```

```python
forecast = pd.DataFrame({'Period': day, 'Short Description': sdes, 'Temperature': temp, 'Description': desc})
forecast.set_index(['Period'], drop= True, inplace=True)
forecast
```
